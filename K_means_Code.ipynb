```python
#Importing relevant libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Import Data นำเข้าข้อมูลจาก Github
data = pd.read_csv('https://github.com/Niyada0309/K-means/raw/main/500_Person_Gender_Height_Weight_Index.csv')
data
```

```python
#Plotting the data ทำการพล็ตข้อมูล
#กำหนดแกน x เป็นน้ำหนัก (Weight) กำหนดแกน y เป็นส่วนสูง (Height)
plt.scatter(data['Weight'],data['Height'])
#กำหนดให้แกน x คือน้ำหนักตั้งแต่ 40-150 kg
plt.xlim(40,150)
#กำหนดให้แกน y คือส่วนสูงตั้งแต่ 100-200 cm
plt.ylim(100,200)
plt.show()
```

```python
#Selecting the feature เลือกคุณสมบัติ
#เลือกข้อมูล row 1 = ส่วนสูง (Height) และ row 2 = น้ำหนัก (Weight)
x = data.iloc[:,1:3]
x
```

```python
#Clustering ทำการจัดกลุ่มออกเป็น 6 กลุ่มตามดัชนี (Index)
kmeans = KMeans(6)
kmeans.fit(x)
```

```python
#Clustering Results of array จัดกลุ่มข้อมูลออกมาเป็น array
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
#Clustering Results แสดงผลลัพธ์การจัดกลุ่ม
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
#แสดงผลออกมาเป็นรูปภาพ โดยให้แกน x = น้ำหนัก และ แกน y = ส่วนสูง
plt.scatter(data_with_clusters['Weight'],data_with_clusters['Height'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
